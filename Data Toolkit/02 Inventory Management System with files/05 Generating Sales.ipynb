#  Product Details:

### 1,Product ID
### 2,Product Name
### 3,Product Price 
### 4,Product Quantity

```python
# Noting the Time of transaction
import time

# Reading the Inventory
fd=open('05 inven.txt','r')
products = fd.read().split('\n')
fd.close()

# Taking user input
ui_prod_username = input('Enter your Name         :')
ui_prod_phone    = input('Enter your Phone No.    :')
ui_prod_mail     = input('Enter your Mail ID      :')
ui_prod_id       = input("Enter Product ID        :")
ui_prod_qn       = input("Enter Product Quantity  :")

updated_prod_lst = []

# Going through each product detail
for product in products:
    
    prod_details = product.split(',')
    
    if(ui_prod_id == prod_details[0]):
        # Checking if product exist or not
        
        # If Product is out of stock
        if(prod_details[3]  ==  '0'):
            print("Sorry we not having this Product.")
            print("This Product is out of Stock.")
            print("Thanks")
            
        else:
            
                if(int(ui_prod_qn) <=  int(prod_details[3])):
                    # If we are having enough Quantity
            
                    print("----------------------------------")
                    print("Product Name   :",prod_details[1])
                    print("Product Price  :",prod_details[2])
                    print("Quantity       :",ui_prod_qn)
                    print("----------------------------------")
                    print("Billing Amount :",int(ui_prod_qn)*int(prod_details[2]))
                    print("----------------------------------")
                    
                    # Updating Inventory List
                    prod_details[3] = str(int(prod_details[3]) - int(ui_prod_qn))
                    
                    # Generating Sales in sales.txt
                    fd = open('05 sales.txt','a')
                    sales_details = ui_prod_username +','+ ui_prod_phone +','+ ui_prod_mail +','+ ui_prod_id +','+ prod_details[1] +','+ ui_prod_qn +','+ str(int(ui_prod_qn)*int(prod_details[2])) +','+ time.ctime() +'\n'
                    fd.write(sales_details)
                    fd.close()
            
                else:
                    # If we are not having enough Quantity
                    
            
                    print("Sorry we are not having enough quantity.")
                    print("We are only having",prod_details[3],"quantity.")
                    print("Would you like to purchase it?")
            
                    ch = input(" Enter 'Y' or 'N' ")
            
            
                    if(ch == 'Y' or ch =='y'):
                    # If you want to purchase with remaining Quantity
            
            
                        print("----------------------------------")
                        print("Product Name   :",prod_details[1])
                        print("Product Price  :",prod_details[2])
                        print("Quantity       :",prod_details[3])
                        print("----------------------------------")
                        print("Billing Amount :",int(prod_details[3])*int(prod_details[2]))
                        print("----------------------------------")
                        
                        # Generating Sales in sales.txt
                        fd = open('05 sales.txt','a')
                        sales_details = ui_prod_username +','+ ui_prod_phone +','+ ui_prod_mail +','+ ui_prod_id +','+ prod_details[1] +','+ prod_details[3] +','+ str(int(prod_details[3])*int(prod_details[2])) +','+ time.ctime() +'\n'
                        fd.write(sales_details)
                        fd.close()
                        
                        # Updating Inventory List
                        prod_details[3] = '0'
            
                    else:
                        print('Thanks')
            
            
    # Updating my Inventory List    
    updated_prod_lst.append(prod_details)
    
lst = []


# Updating my Inventory String
for i in updated_prod_lst:
    
    prod = i[0]+','+ i[1] + ',' + i[2] + ',' + i[3] + '\n'
    
    lst.append(prod)

    
# Removing last \n from the list    
lst[-1] = lst[-1][:-1]


# Updating Inventory file
fd = open('05 inven.txt','w')

for i in lst:
    fd.write(i)
fd.close()

print("-------------------")
print("Inventory Updated")
```
